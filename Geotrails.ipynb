```python
%%time
import gpxpy
import gpxpy.gpx

import argparse

import pandas as pd
import numpy as np

import datashader as ds
import datashader.glyphs
import datashader.transfer_functions as tf

from os import listdir
from os.path import isfile, join

import pyproj

gpx_file_path = 'gpx_files'
gpx_files = [f for f in listdir(gpx_file_path) if isfile(join(gpx_file_path, f)) and f.endswith('.gpx')]

points = []

in_proj = pyproj.Proj(init='epsg:4326') # WGS84
out_proj = pyproj.Proj(init='epsg:3857') # Web Mercator


for gpx_file in gpx_files:
    print("Processing {0}".format(gpx_file))
    with open(join(gpx_file_path, gpx_file), 'r') as f:
        try:
            gpx = gpxpy.parse(f)

            for track in gpx.tracks:
                for segment in track.segments:
                    for point in segment.points:
                        x,y = pyproj.transform(in_proj, out_proj, point.longitude, point.latitude)
                        points.append({'x': x, 
                                       'y': y, 
                                       'lat': point.latitude, 
                                       'lon': point.longitude, 
                                       'elev': point.elevation, 
                                       'speed': point.speed })

        except (UnicodeDecodeError, gpxpy.parser.mod_gpx.GPXXMLSyntaxException):
            print("Failed to process {0}".format(gpx_file))

print("Processed {0} points".format(len(points)))

df = pd.DataFrame(points)
```

```python
#%%time
import bokeh.plotting as bp
from bokeh.tile_providers import STAMEN_TERRAIN
from datashader.bokeh_ext import InteractiveImage

bp.output_notebook()
#p = bp.figure(tools='pan,wheel_zoom,reset', x_range=(-180,180), y_range=(-90,90))
p = bp.figure(tools='pan,wheel_zoom,reset', x_range=(-1800000,1800000), y_range=(-900000,900000))
p.add_tile(STAMEN_TERRAIN)

def image_callback(x_range, y_range, w, h):
    cvs = ds.Canvas(plot_width=w, plot_height=h, x_range=x_range, y_range=y_range)
    #agg = cvs.points(df, 'lon', 'lat', ds.mean('elev'))
    agg = cvs.points(df, 'x', 'y', ds.mean('elev'))
    img = tf.shade(agg)
    return tf.dynspread(img, threshold=0.50)

InteractiveImage(p, image_callback)

```
